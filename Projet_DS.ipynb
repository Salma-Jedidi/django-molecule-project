```
def smiles_to_fingerprints(smiles_list, fp_size=2048):
    fps = []
    for smiles in smiles_list:
        mol = Chem.MolFromSmiles(smiles)
        if mol is not None:
            fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=fp_size)
            fp_array = np.zeros((1,), dtype=int)
            DataStructs.ConvertToNumpyArray(fp, fp_array)
            fps.append(fp_array)
    return np.array(fps)

# Apply the conversion to the dataset
df_test['Fingerprint'] = df_test['Smiles'].apply(smiles_to_fingerprint)

# Remove rows with failed fingerprint conversion
df_test = df_test.dropna(subset=['Fingerprint'])

# Reset the index
df_test = df_test.reset_index(drop=True)
```

```
import gc

# Define a function to process a chunk of data
def process_chunk(chunk_df):
    chunk_df['rdkit_mol'] = chunk_df['Smiles'].apply(MolFromSmiles)
    return chunk_df

# Define the chunk size
chunk_size = 500

# Create an empty DataFrame to store the processed data
processed_df = pd.DataFrame()

# Process the dataset in chunks
for i in range(0, chembl_df_test.shape[0], chunk_size):
    chunk = chembl_df_test.iloc[i:i + chunk_size]
    processed_chunk = process_chunk(chunk)
    processed_df = processed_df.append(processed_chunk, ignore_index=True)

    # Free up memory
    del chunk
    del processed_chunk
    gc.collect()

# Replace the original dataset with the processed dataset
chembl_df_test = processed_df
```
